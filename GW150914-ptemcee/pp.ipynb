```python
import arviz as az
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns

sns.set_palette('husl')
sns.set_context('notebook')
```

```python
trace = az.from_netcdf('checkpoint.nc')
```

```python
fig, sg = plt.subplots(5, 4)
for i, k in enumerate(trace.posterior.keys()):
    ii = i // 4
    jj = i % 4
    
    sg[ii,jj].plot(trace.posterior[k].mean(dim='chain'))
```
